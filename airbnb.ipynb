```python
#Import all Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Loading the datasets
data=pd.read_csv('datasets.csv',encoding_errors='ignore')
```

```python
#intial exploration
data.head()
```

```python
data.info()
```

```python
#Statiscal summary of the data
data.describe()
```

```python
#Data Cleaning(Missing values)
data.isnull().sum()
```

```python
# Droping the rows with missing values
data.dropna(inplace=True)
data.isnull().sum()
```

```python
#Dealing with duplicate Rows

data.duplicated().sum()
data[data.duplicated()]
```

```python
#Deleting the duplicate rows
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
#Data inconsistencies
data.dtypes
```

```python
# Type Casting (Change the data type of id and host_id to object)

data['id']= data['id'].astype('object')
data['host_id']= data['host_id'].astype('object')
data.dtypes
```

# EDA
##  Data Analysis and Visualization
### 1. Univariate Analysis ( Analysis of individual columns and their distributions)
### 2. Bivariate Analysis (Analysis of relationship between two or more columns or effect of one column on another column)



```python
#Price distribution
data['price']
```

```python
# Identifying outliers using box plot
sns.boxplot(data['price'])
```

```python
# Identifying  outliers in price column
df = data[data['price'] < 1500]
sns.boxplot(data=df['price'])

```

```python
# Price distribution after removing outliers
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='price', bins=100, kde=True)
plt.title('Price Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
data.columns
```

```python
# Price distribution
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='availability_365')
plt.title('availability_365 Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
# Find neighbourhood group with average price
df.groupby(by='neighbourhood_group')['price'].mean()
```

```python
# Feature Engineering
# Create a new Column 'price_per_bedroom'
df['price_per_bed'] = df['price'] / (df['beds'])  # Adding 1 to avoid division by zero
df.head()
```

```python
#  average price per bedroom in each neighbourhood group
df.groupby(by='neighbourhood_group')['price_per_bed'].mean()
```

### Bivariate Analysis

```python
# How the  price is dependent on the neighbourhood group and room type
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='neighbourhood_group', y='price', hue='room_type')
plt.title('Price vs Neighbourhood Group')
plt.ylabel('Price')
plt.show()
```

```python
# Number of reviews and price relationship
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='number_of_reviews', y='price',hue='neighbourhood_group')
plt.title('Number of Reviews vs Price')
plt.ylabel('Price')
plt.show()
```

```python
sns.pairplot(data=df, vars=['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365'], hue='neighbourhood_group')
```

```python
# Geographical Analysis airbnb
sns.scatterplot(data=df, x='longitude', y='latitude', hue='neighbourhood_group')
```

```python
# Correlation heatmap correlation one variable with other numerical columns
corr=df[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']].corr()
corr
plt.figure(figsize=(8,5))
sns.heatmap(corr, annot=True)
```
